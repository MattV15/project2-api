```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Load the dataset from a CSV file
df = pd.read_csv('pudding_data.csv')  # Replace with your actual CSV file name

# Create a new column to store the downloaded text
df['script_text'] = None

# Loop through the links and download the first 1000 characters
for index, row in df.iterrows():
    link = row['link']
    # Check if link is a valid URL
    if link.startswith('http'):
        try:
            response = requests.get(link, timeout=10)
            if response.status_code == 200:
                # Parse the HTML content
                soup = BeautifulSoup(response.text, 'html.parser')
                
                # Get the text and remove HTML tags
                script_text = soup.get_text()
                
                # Get the first 1000 characters
                script_text = script_text[:1000]

                # Save the script text in the DataFrame
                df.at[index, 'script_text'] = script_text

                print(f'Downloaded script for index {index} from {link}')
            else:
                print(f'Failed to retrieve {link}: Status code {response.status_code}')
        except requests.exceptions.RequestException as e:
            print(f'Error fetching {link}: {e}')
    else:
        print(f'Invalid URL: {link}')

# Save the updated DataFrame to a new CSV file
df.to_csv('pudding_texts.csv', index=False, encoding='utf-8')
print('Saved the DataFrame with downloaded texts to pudding_texts.csv')
```

```python
import requests
import pandas as pd

# Set up the TMDb API URL and headers
api_key = "0b960ee3126762d5af1c0b772a8caf4a"
url = "https://api.themoviedb.org/3/movie/now_playing"

# Set up parameters
params = {
    "api_key": api_key,
    "language": "en-US",
    "page": 1  # You can change this to get more pages if needed
}

# Make the API request
response = requests.get(url, params=params)

# Check if the response is successful
if response.status_code == 200:
    data = response.json()
    
    # Extract relevant movie information
    movies = []
    for movie in data['results']:
        movie_info = {
            "title": movie.get("title"),
            "release_date": movie.get("release_date"),
            "overview": movie.get("overview"),
            "vote_average": movie.get("vote_average"),
            "poster_path": f"https://image.tmdb.org/t/p/w500{movie.get('poster_path')}",
            "original_language": movie.get("original_language"),
            "popularity": movie.get("popularity"),
            "id": movie.get("id"),
        }
        movies.append(movie_info)

    # Create a DataFrame
    df = pd.DataFrame(movies)

    # Save the dataset to a CSV file
    df.to_csv('now_playing_movies.csv', index=False)
    print("Dataset created and saved as 'now_playing_movies.csv'.")

else:
    print(f"Failed to retrieve data: {response.status_code} - {response.text}")
```
